# 🔷 Why determinant works only for square matrices

### 1. Determinant measures **volume scaling**

A determinant tells you:

> “How much does this matrix stretch or shrink space?”

* 2×2 matrix → area scaling
* 3×3 matrix → volume scaling
* n×n matrix → n-dimensional volume scaling

👉 This only makes sense when:

* Input space and output space have the **same dimension**

---

### ❌ What happens for non-square matrices?

If matrix is:

* **m × n where m ≠ n**

Then it maps:

* ℝⁿ → ℝᵐ (different dimensions)

Example:

* 2×3 matrix → maps 3D → 2D
* You’re **flattening space**, not scaling it

👉 So:

* Volume is not preserved
* You **cannot define a single scaling factor**

➡️ Hence, **determinant is undefined**

---

# 🔷 Why inverse exists only for square matrices

### 2. Inverse means “undo the transformation”

If matrix **A has an inverse**, then:

[
A^{-1} A = I
]

This means:

* Whatever A does, A⁻¹ reverses it perfectly

---

### ❌ For non-square matrices

Suppose A is **m × n**

* Forward: ℝⁿ → ℝᵐ
* Reverse would need: ℝᵐ → ℝⁿ

👉 But:

* If dimensions differ, you **lose or gain information**

---

### 🔍 Two cases

#### Case 1: m < n (compression)

Example: 2×3 matrix

* 3D → 2D → information lost
* Cannot recover original input

➡️ No inverse possible

---

#### Case 2: m > n (expansion)

Example: 3×2 matrix

* 2D → 3D
* Infinite inputs can map to same output

➡️ No unique inverse

---

# 🔷 Intuition (very important)

| Concept     | Requirement   | Why                             |
| ----------- | ------------- | ------------------------------- |
| Determinant | Square matrix | Same dimension → volume scaling |
| Inverse     | Square matrix | Must map back uniquely          |

---

# 🔷 Bonus (advanced insight — useful for you)

Even though inverse doesn’t exist for non-square matrices, we use:

👉 **Pseudo-inverse (Moore–Penrose inverse)**
Used in:

* Machine learning
* Regression (OLS solution)

---

# 🔥 One-line takeaway

> Square matrices are the only ones that transform a space **into itself**, which is why both determinant and inverse are defined only for them.

# Softmax (dim=0) — Mathematical Explanation

## Given Matrix

k = [ [1, 2],
[3, 4] ]

---

## Softmax Formula

For any element x_i:

softmax(x_i) = exp(x_i) / sum_j exp(x_j)

---

## Applying softmax with dim = 0

This means we apply softmax column-wise (down each column).

---

## Column 1: [1, 3]

Step 1: Exponentials
exp(1) = 2.718
exp(3) = 20.085

Step 2: Sum
2.718 + 20.085 = 22.803

Step 3: Normalize
2.718 / 22.803 = 0.1192
20.085 / 22.803 = 0.8808

---

## Column 2: [2, 4]

Step 1: Exponentials
exp(2) = 7.389
exp(4) = 54.598

Step 2: Sum
7.389 + 54.598 = 61.987

Step 3: Normalize
7.389 / 61.987 = 0.1192
54.598 / 61.987 = 0.8808

---

## Final Output Matrix

softmax(k, dim=0) =

[ [0.1192, 0.1192],
[0.8808, 0.8808] ]

---

## General Formula for dim = 0

softmax(k_ij) = exp(k_ij) / sum_i exp(k_ij)

Where:

* i = row index
* j = column index

So normalization happens over rows (vertical direction).

---

## Key Insight

* dim = 0 → normalize each column
* Each column sums to 1

Column sums:
0.1192 + 0.8808 = 1

---

## One-Line Summary

softmax(k, dim=0) converts each column into a probability distribution.

## CPU vs GPU

```python
import torch
import time

# Define the size of the matrices
size = 10000  # Large size for performance comparison

# Create random matrices on CPU
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# -----------------------------
# Measure time on CPU
# -----------------------------
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)
cpu_time = time.time() - start_time

print(f"Time on CPU: {cpu_time:.4f} seconds")

# -----------------------------
# Move matrices to GPU
# -----------------------------
if torch.cuda.is_available():
    matrix_gpu1 = matrix_cpu1.to('cuda')
    matrix_gpu2 = matrix_cpu2.to('cuda')

    # -----------------------------
    # Measure time on GPU
    # -----------------------------
    torch.cuda.synchronize()  # Ensure clean start
    start_time = time.time()

    result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)

    torch.cuda.synchronize()  # Wait for GPU to finish
    gpu_time = time.time() - start_time

    print(f"Time on GPU: {gpu_time:.4f} seconds")

else:
    print("CUDA (GPU) is not available on this system.")
```
