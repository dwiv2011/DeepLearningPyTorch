Autograd automatically computes gradients by tracking operations and applying the chain rule, which is essential for training ML models.

```python
import torch
```

```python
x=torch.tensor(3.0,requires_grad=True)# requires_grad is required when we want to compute gradient on the derived tensor
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

Creating AutoGrad

```python
#Autograd creates a DAG to capture the sequence
```

```python
x=torch.tensor(3.0,requires_grad=True) # require_grad is require at the leaf node
y=x**2 # Y is called intermediate Node
z=2*y+3#Z is root node
```

```python
#now to calculate grad we need to initiate the call.
z.backward()
```

```python
x.grad
```

```python
# if we calculate the grad at the intermoediate tensor
y.grad
```

```python
#problem with grad  it accumulates the grad by adding the value in each call.
# to avoid, we need to flush the gradient value before each call
```

```python
y=x**2 # Y is called intermediate Node
z=2*y+3#Z is root node
```

```python
z.backward()
```

```python
x.grad # now it becme 24 , earlier it was 12
```

```python
#to flush the value
x.grad.zero_()
```

```python
# to turn off the gradient tracking, take a scenario where model has trained and
#now we have to predict if we dont turn off then it will be tracking the gradient and consuming lost of memory while predicting as well.
```

```python
#Option 1:
x.requires_grad_(False)
```

```python
#option 2
# detaching the grad functionality and assigning to y, now if we calll variable y and that does not have grad
y=x.detach()
```

```python
with torch.no_grad():
  y=x**2
```

```python
y
```

```python
y.backward() #because tracking is turned off
```
